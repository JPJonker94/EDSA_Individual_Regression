```python
#Joined df_train and df_test data to ensure updating of features is done on both datasets.
df = pd.concat([df_train,df_test])
df.shape
```

```python
# Updated missing values/ features
# Valencia_pressure missing data was updated for both data sets
# Mode was used to update the missing values. This updated produced the best RMSE for the different models experimented with.
print(f"\n mode: {df['Valencia_pressure'].mode()[0]} \n median: {df['Valencia_pressure'].median()} \n mean: {df['Valencia_pressure'].mean()}")
df['Valencia_pressure'] = df['Valencia_pressure'].fillna(df['Valencia_pressure'].mode()[0])

#Confirm that all values was updated
df['Valencia_pressure'].isnull().sum()
```

```python
#Change order of columns and show response variable as last column
column_order = ['Madrid_wind_speed', 'Valencia_wind_deg', 'Bilbao_rain_1h',
       'Valencia_wind_speed', 'Seville_humidity', 'Madrid_humidity',
       'Bilbao_clouds_all', 'Bilbao_wind_speed', 'Seville_clouds_all',
       'Bilbao_wind_deg', 'Barcelona_wind_speed', 'Barcelona_wind_deg',
       'Madrid_clouds_all', 'Seville_wind_speed', 'Barcelona_rain_1h',
       'Seville_pressure', 'Seville_rain_1h', 'Bilbao_snow_3h',
       'Barcelona_pressure', 'Seville_rain_3h', 'Madrid_rain_1h',
       'Barcelona_rain_3h', 'Valencia_snow_3h', 'Madrid_weather_id',
       'Barcelona_weather_id', 'Bilbao_pressure', 'Seville_weather_id',
       'Valencia_pressure', 'Seville_temp_max', 'Madrid_pressure',
       'Valencia_temp_max', 'Valencia_temp', 'Bilbao_weather_id',
       'Seville_temp', 'Valencia_humidity', 'Valencia_temp_min',
       'Barcelona_temp_max', 'Madrid_temp_max', 'Barcelona_temp',
       'Bilbao_temp_min', 'Bilbao_temp', 'Barcelona_temp_min',
       'Bilbao_temp_max', 'Seville_temp_min', 'Madrid_temp', 'Madrid_temp_min', 
       'year', 'month', 'day', 'hour', 'weekday', 'week',
       'load_shortfall_3h']
df = df[column_order]

#Split data according to the train and test data after data cleaning
train_data = df.iloc[df_train.index]
test_data = df.iloc[max(df_train.index)+1:]

#Remove NaN response variable from testing data
test_data = test_data.drop(['load_shortfall_3h'], axis=1)

#Show new data split columns
print(train_data.shape, test_data.shape)
```

```python
#View test data to confirm removal of load_shortfall_3h columns
test_data.head()
```

```python
# create targets and features dataset from train_data
X = train_data.drop(['load_shortfall_3h'],axis=1)
Y = train_data['load_shortfall_3h']

#Split training data into training and testing datasets
X_train, x_test, y_train, y_test = train_test_split(X, Y , test_size=0.2,random_state=PARAMETER_CONSTANT, shuffle=True)
```

```python
# create one or more ML models
# Create a decision tree with multiple max_depth to determine best max_depth for ensemble models

#Empty variables to record the results from model iteration
recording = []
depth = []

#For loop with linear spaced numbers
for i in np.linspace(1,20,20):
    #Created a DecisionTree instance
    dt = DecisionTreeRegressor(max_depth=int(i))
    #Fit model with training data
    dt.fit(X_train,y_train)
    #Predict the response for training data
    y_predict = dt.predict(X_train)
    #Predict the response for testing data
    y_test_predict = dt.predict(x_test)
    
    #Append the RMSE results to the empty variables
    recording.append([int(i),math.sqrt(mean_squared_error(y_train,y_predict)),math.sqrt(mean_squared_error(y_test,y_test_predict))])
    depth.append(int(i))

#Created Dataframe to show max_depth RMSE results    
result_df = pd.DataFrame(recording, columns=['depth', 'Train_RMSE', 'Test_RMSE'])
result_df['difference'] = result_df['Test_RMSE'] - result_df['Train_RMSE']
result_df
```

```python
#Created figure to show differce of RMSE at each max_depth for DecisionTreeRegressor
plt.figure(figsize=(20,20))
plt.xticks(depth)
plt.plot(depth, result_df['Train_RMSE'])
plt.plot(depth, result_df['Test_RMSE'], color='red')
```

```python
#Creating list of model names to enable multiple model result generating
names = ['LinearRegression','Ridge','Lasso',
         'DecisionTreeRegressor','BaggingRegressor',
         'AdaBoostRegressor','RandomForestRegressor','XGBoost']

#Creating list of instances for different models and implementing max_depth based on previous results 
regression = [LinearRegression(),Ridge(),Lasso(),DecisionTreeRegressor(max_depth=6),BaggingRegressor(base_estimator=DecisionTreeRegressor(max_depth=6)),
         AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=6)),RandomForestRegressor(), xgb.XGBRegressor()]
```

```python
#Empty results variable to capture all of the RMSE information for the different models
results = []

#Creating a zip object containing a tuples of combination of model name and instance of model 
for name, instance in zip(names,regression):
    print(f'{name}')
    instance.fit(X_train, y_train)
    print(f'Predicting: Train data for {name}')
    y_train_pred = instance.predict(X_train)
    rmse_y_train = math.sqrt(mean_squared_error(y_train, y_train_pred))
    print(f'Predicting: Test data for {name}\n')
    y_test_pred = instance.predict(x_test)
    rmse_y_test = math.sqrt(mean_squared_error(y_test, y_test_pred))
    #Append resuts to empty result list
    results.append([name,rmse_y_train,rmse_y_test])  
```

```python
#Show result before loading to a Dataframe
results
```

```python
#Implement same process for Heterogeneous models

#Create instances for model estimators
lin_reg = LinearRegression()
destree = DecisionTreeRegressor(max_depth=6)
lass = Lasso()

#Create tuple combinations of model name and instance
estimator = [("LR",lin_reg),("LS",lass),("DT",destree)]

#Determine meta learner for Stacking regressor
meta_learner_reg = LinearRegression()
#Determine weighted average of each estimator model
model_weightings = np.array([0.1,0.1,0.8])

#List of Heterogeneous models names
names_hete = ['VotingRegressor','StackingRegressor']
#List of Heterogeneous models instances
model_hete = [VotingRegressor(estimators=estimator,weights=model_weightings),
              StackingRegressor(estimators=estimator, final_estimator=meta_learner_reg)]

#Creating a zip object containing a tuples of combination of model name and instance of model 
for name, reg in zip(names_hete,model_hete):
    print(f'{name}')
    reg.fit(X_train, y_train)
    print(f'Predicting: Train data for {name}')
    y_train_pred = reg.predict(X_train)
    rmse_y_train = math.sqrt(mean_squared_error(y_train, y_train_pred))
    print(f'Predicting: Test data for {name}\n')
    y_test_pred = reg.predict(x_test)
    rmse_y_test = math.sqrt(mean_squared_error(y_test, y_test_pred))
    
    #append results to result list
    results.append([name,rmse_y_train,rmse_y_test])

#Create new result dataframe for all models
results = pd.DataFrame(results, columns=['Model','Train_RMSE', 'Test_RMSE'])
results['difference'] = results['Test_RMSE'] - results['Train_RMSE']
results.sort_values('Test_RMSE')
```

AdaBoostRegressor Trial with iteration

```python
#To ensure no data was manipulated I will call a new train test dataset for each model tested
#Spesify X and Y training data
#Remove 'load_shortfall_3h' from x dataset
X_train_ada = train_data.drop(['load_shortfall_3h'],axis=1)
y_train_ada = train_data['load_shortfall_3h']

#Spesify X and Y testing data
test_data = df.iloc[max(df_train.index)+1:]
#Remove 'load_shortfall_3h' from x dataset 
x_test_ada = test_data.drop(['load_shortfall_3h'], axis=1)

#Create standardisation instance
scaler = StandardScaler()
#Standadise x_train data
X_scaled = scaler.fit_transform(X_train_ada)
#Standadise x_test data
X_test_scaled =scaler.fit_transform(x_test_ada)

#Empty list to capture iteration of different adamodels
model_AdaBoost_list = []
#Empty list to capture iteration of different adamodels results
results_AdaBoost_model = []

#Foor loop to run 100 different models on AdaBoost to determine the best performing model
for i in np.linspace(1,100,100):
    #Spesify train test split based on training data
    X_train, x_test, y_train, y_test = train_test_split(X_scaled, y_train_ada, test_size=0.2)
    #Create AdaBoostRegressor instance
    ada_model = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=8),random_state=PARAMETER_CONSTANT)
    #Fit training data
    ada_model.fit(X_train, y_train)
    #Predict x_train data
    y_train_ada_pred = ada_model.predict(X_train)
    #Predict x_test data
    y_test_ada_pred = ada_model.predict(x_test)
    #Append model to model list
    model_AdaBoost_list.append(ada_model)
    
    #Calculate results to results_AdaBoost_model
    result = [f'ada_model {int(i)}', math.sqrt(mean_squared_error(y_train,y_train_ada_pred)),math.sqrt(mean_squared_error(y_test,y_test_ada_pred))]
    #Add results to results_AdaBoost_model
    results_AdaBoost_model.append(result)
    #print model iteration
    print(int(i))
```

```python
#Add all of the iteration results to a pandas DataFrame called df_results_AdaBoost_model
df_results_AdaBoost_model = pd.DataFrame(results_AdaBoost_model, columns=['Model', 'Train_RMSE', 'Test_RMSE'])
#Calculate the difference between Train RMSE and Test RMSE
df_results_AdaBoost_model['difference'] = df_results_AdaBoost_model['Test_RMSE'] - df_results_AdaBoost_model['Train_RMSE']

#Check for the best performing Adaboost model in dataframe
print(fr"Best model was ada_model: {df_results_AdaBoost_model[df_results_AdaBoost_model['difference']==df_results_AdaBoost_model['difference'].min()].index[0]+1}")
#Print best performing Adaboost model RMSE inforamtion
print(df_results_AdaBoost_model[df_results_AdaBoost_model['difference']==df_results_AdaBoost_model['difference'].min()])

#Export all Adaboost model iteration information to excel
df_results_AdaBoost_model.to_excel('AdaBoost_model_iterations.xlsx')

#Create final dataframe of best Adaboost model
final_df_results_AdaBoost_model = df_results_AdaBoost_model[df_results_AdaBoost_model['difference']==df_results_AdaBoost_model['difference'].min()]
```

```python
#Spesify training data
train_data = df.iloc[df_train.index]
#Spesify training data x and y
X_train_rf = train_data.drop(['load_shortfall_3h'],axis=1)
y_train_rf = train_data['load_shortfall_3h']

#Standardise the X train data
Scaler = StandardScaler()
X_scaled = Scaler.fit_transform(X_train_rf)

#Create train test split for RandomForestRegressor iteration
X_train, x_test, y_train, y_test = train_test_split(X_scaled,y_train_rf,test_size=0.2, shuffle=True)

#Empty list to save all of RandomForest models
RFR_models_results = []
RFR_models_save = []

#iteration to calculate the best random forest output
for i in np.linspace(1,10,10):
    #Random forest instance
    RFR = RandomForestRegressor(n_estimators=100, max_depth=6)
    #Random forest training
    RFR.fit(X_train,y_train)
    #Predixt x train
    rf_y_train = RFR.predict(X_train)
    #Predict x Test
    rf_y_test = RFR.predict(x_test)
    #Save results to list
    RFR_models_results.append(['RFR_iteration',math.sqrt(mean_squared_error(y_train, rf_y_train)), math.sqrt(mean_squared_error(y_test, rf_y_test))])
    RFR_models_save.append(RFR)

#Create dataframe to safe results
rf_df = pd.DataFrame(RFR_models_results, columns=['Model', 'Train_RMSE', 'Test_RMSE'])
rf_df['difference'] = rf_df['Test_RMSE'] - rf_df['Train_RMSE']
rf_df
```

```python
#Combine result data for all models
all_model_results = pd.concat([all_model_results,rf_df[rf_df['difference']==rf_df['difference'].min()]])
#Display best performing models
all_model_results.sort_values('Test_RMSE')
```

```python
#Assign test data
test_data = df.iloc[max(df_train.index)+1:]
#Drop load_shortfall_3h from test data
x_test_rf = test_data.drop(['load_shortfall_3h'], axis=1)

#Scale x test data
Scaler = StandardScaler()
X_test_scaled =Scaler.fit_transform(x_test_rf)
#Create randomforest instance
RFR = RandomForestRegressor(n_estimators=100, max_depth=6)
#Fit with all data
RFR.fit(X_scaled,y_train_rf)
#Predict X_test_scaled
y_test_pred = RFR.predict(X_test_scaled)

#Create df for prediction data
prediction = pd.DataFrame(y_test_pred, columns=['load_shortfall_3h'])
#Extract time from test df
submission = pd.DataFrame({'time':df_test['time']})
#Combine test prediction and time df
submission_final = submission.join(prediction)
#Export data to csv
submission_final.to_csv("submission_RF.csv", index=False)
```
